## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df["line_subtotal"] = df["unit_price"] * df["qty"]
df.head()
```

```python
# Create a column for shipping price.Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df["shipping_price"] = df.apply(lambda row: row["unit_weight"] * row["qty"] * 7 if row["unit_weight"] * row["qty"] > 50 else row["unit_weight"] * row["qty"] * 10, axis=1)
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["total_price"] = (df["line_subtotal"] + df["shipping_price"] ) * 1.0925
df.head(10)
```

```python
# Create a column for the cost of each line using unit cost, qty, and shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df["line_profit"] = df["total_price"] - df["line_cost"]
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
def client_spend(client_id, total_price):
    client_df = df.loc[df['client_id'] == client_id, total_price]
    return round(client_df.sum(), 2)
for client_id in top_client_ids:
    print(f"{client_id}: ${client_spend(client_id, 'total_price')}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
data_dict = {"client_id": top_client_ids}
columns = ["qty", "shipping_price", "total_price", "line_cost", "line_profit"]
for column in columns:
    data_dict[column] = [client_spend(client_id, column) for client_id in top_client_ids]

summary_df = pd.DataFrame(data_dict).sort_values("line_profit", ascending=False)
summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
money_columns = ['line_cost', 'shipping_price', 'total_price', 'line_profit']

def format_money(amount):
    return f"${amount/1000000:.2f}M"

for column in money_columns:
    summary_df[column] = summary_df[column].apply(format_money)

rename_columns = {
    'qty': 'Total Units Purchased',
    'shipping_price': 'Total Shipping Price',
    'total_price': 'Total Revenue',
    'line_profit': 'Total Profit',
    'client_id': 'Client ID',
    'line_cost': 'Total Cost'
}
summary_df = summary_df.rename(columns=rename_columns)
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_df.sort_values('Total Profit', ascending=False)
```
